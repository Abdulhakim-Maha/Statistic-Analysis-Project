```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ['แรงม้า','เงินทุนรวม','คนงานรวม']
dataset = pd.read_csv('hp_no_outlier.csv',encoding='utf-8',usecols=cols)
dataset
```

```python
dataset.describe()
```

```python
dataset.mode()
```

```python
Q1_fund = dataset['เงินทุนรวม'].quantile(0.25)
Q3_fund = dataset['เงินทุนรวม'].quantile(0.75)
Q1_fund,Q3_fund
```

```python
IQR_fund = Q3_fund - Q1_fund
IQR_fund
```

```python
lower_outlier_fund = Q1_fund - 1.5*IQR_fund
upper_outlier_fund = Q3_fund + 1.5*IQR_fund
lower_outlier_fund, upper_outlier_fund
```

```python
dataset[(dataset['เงินทุนรวม'] < lower_outlier_fund)|(dataset['เงินทุนรวม'] > upper_outlier_fund)]
```

```python
dataset[(dataset['เงินทุนรวม'] > lower_outlier_fund) & (dataset['เงินทุนรวม'] < upper_outlier_fund)]
```

```python
fund_no_outlier = dataset[(dataset['เงินทุนรวม'] > lower_outlier_fund) & (dataset['เงินทุนรวม'] < upper_outlier_fund)]
fund_no_outlier
```

```python
fund_no_outlier.describe()
```

```python
fund_no_outlier.mode()
```

```python
fund_no_outlier.to_csv(r'./fund_no_outlier.csv')
```
